```python
tocompare = ['US', 'Italy']
l = df2.loc[tocompare[0]][2::].shape[0]
D = np.zeros((l,2))
D[:,0] = np.array(df2.loc[tocompare[0]][2::])
D[:,1] = np.array(df2.loc[tocompare[1]][2::])
Dr = []
for i in range(l):
    if i == 0:
        r1, r2 = 0.0, 0.0
    else:
        r1 = D[i,0] - D[i-1,0]
        r2 = D[i,1] - D[i-1,1]
    Dr.append((r1,r2))
Dr = np.array(Dr)

#f1 = np.polyfit(D[:,0], Dr[:,0], 2)
#f2 = np.polyfit(D[:,1], Dr[:,1], 1)
#print(f1)

plt.figure(figsize=(16,8))
plt.plot(D[:,0], Dr[:,0], 'o', label=tocompare[0])
plt.plot(D[:,1], Dr[:,1], 'x', label=tocompare[1])
plt.xlabel('Confirmed Cases')
plt.ylabel('Rate')
plt.legend()
plt.grid()
#plt.ylim((0,Dr[:,0].max()))
#plt.xlim((0,D[:,0].max()))
```

```python
import scipy.optimize as opt
colors=['b','g','r','c','m','y','k','c']

def func(x,a,b,c):
    return c+b*x+a*x*x

tocompare = ['India', 'Italy', 'Spain', 'Russia', 'Iran', 'Germany', 'France', 'US']
l1 = df2.loc[tocompare[0]][2::].shape[0]
l2 = len(tocompare)
D = np.zeros((l1,l2))
for l in range(l2):
    D[:,l] = np.array(df2.loc[tocompare[l]][2::])

#D[:,0] = np.array(df2.loc[tocompare[0]][2::])
#D[:,1] = np.array(df2.loc[tocompare[1]][2::])
Dr = np.zeros((l1,l2))
for j in range(l2):
    for i in range(l1):
        if i == 0:
            Dr[i,j] = 0.0
        else:
            Dr[i,j] = D[i,j] - D[i-1,j]

op, pcov = [], []
for l in range(l2):
    op1, pcov1 = opt.curve_fit(func, D[:,l], Dr[:,l])
    op.append((op1))

f1 = np.polyfit(D[:,0], Dr[:,0], 2)
f2 = np.polyfit(D[:,1], Dr[:,1], 1)
print(f1, f2)

plt.figure(figsize=(16,8))
for l in range(l2):
    plt.plot(D[:,l], Dr[:,l], 'o', color=colors[l], label=tocompare[l])
    plt.plot(D[:,l], func(D[:,l], *op[l]), '-', color=colors[l])
plt.xlabel('Confirmed Cases')
plt.ylabel('Rate')
plt.legend()
plt.grid()
plt.ylim((0,3.5*Dr[:,0].max()))
plt.xlim((0,3.5*D[:,0].max()))
```

```python
import scipy.optimize as opt

def func(x,a,b,c):
    return a*x*x+b*x+c

tocompare = ['India', 'Korea, South']
l = df2.loc[tocompare[0]][2::].shape[0]
D = np.zeros((l,2))
D[:,0] = np.array(df2.loc[tocompare[0]][2::])
D[:,1] = np.array(df2.loc[tocompare[1]][2::])
Dr = []
for i in range(l):
    if i == 0:
        r1, r2 = 0.0, 0.0
    else:
        r1 = D[i,0] - D[i-1,0]
        r2 = D[i,1] - D[i-1,1]
    Dr.append((r1,r2))
Dr = np.array(Dr)

op1, pcov1 = opt.curve_fit(func, D[:,0], Dr[:,0])
op2, pcov2 = opt.curve_fit(func, D[:,1], Dr[:,1])
print(op1,op2)

#f1 = np.polyfit(D[:,0], Dr[:,0], 2)
#f2 = np.polyfit(D[:,1], Dr[:,1], 2)
#print(f1, f2)

plt.figure(figsize=(16,8))
plt.plot(D[:,0], Dr[:,0], 'o', label=tocompare[0])
plt.plot(D[:,0], func(D[:,0], *op1), label='fit1')
plt.plot(D[:,1], Dr[:,1], 'x', label=tocompare[1])
plt.plot(D[:,1], func(D[:,1], *op2), label='fit2')
plt.xlabel('Confirmed Cases')
plt.ylabel('Rate')
plt.legend()
plt.grid()
#plt.ylim((0,1.1*Dr[:,0].max()))
#plt.xlim((0,1.1*D[:,0].max()))
```
